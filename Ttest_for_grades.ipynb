```python
#question: did the average grade of students in the class, G1 and G2, change?
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#import data
data=pd.read_csv('student-por.csv')
```

```python
data
```

```python
#check null value
data.isnull().sum()
```

```python
data.describe()
```

```python
#draw boxplot for G1 and G2
fig = plt.figure(figsize=(8, 5))

# Adds subplot on position 1
ax = fig.add_subplot(121)
# Adds subplot on position 2
ax2 = fig.add_subplot(122)

ax.boxplot(data.G1)
ax2.boxplot(data.G2)
plt.show()
```

```python
G1=data.G1
G2=data.G2
```

```python
#t-test. A p-value of 0.05 or lower is generally considered statistically significant.
t_value,p_value=stats.ttest_ind(G1,G2)
t_value,p_value
```


The average grade of students has improved, but it is not significant
